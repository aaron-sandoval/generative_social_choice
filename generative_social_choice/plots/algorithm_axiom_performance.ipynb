## Load data

```python
from pathlib import Path

data_dir_path = get_base_dir_path() / "data" / "voting_algorithm_evals"
use_latest_data_file = True

if use_latest_data_file:
    # Find the most recently created CSV file
    data_files = list(data_dir_path.glob("*.csv"))
    data_file = max(data_files, key=lambda p: p.stat().st_mtime) if data_files else None
else:
    # Use the hard-coded file
    data_file = data_dir_path / "2025-03-04-204044.csv"
data_file
```

```python

# Load the data with the correct MultiIndex structure
df = pd.read_csv(data_file, index_col=0, header=[0,1])
df
```

# Analysis

```python
# Filter to only include columns where there are at least 2 unique values
# This removes axioms that all algorithms either pass or fail
filtered_df = df.loc[:, df.nunique() >= 2]
print(f"{filtered_df.shape=}")
print(f"{filtered_df.shape[1]/df.shape[1]=}")
```

```python

# Display the filtered dataframe
filtered_df
```

Quick overall assessment of voting alg performance on differential axiom-cases

```python
filtered_df.mean(axis=1)
```

```python
# Group algorithms with identical axiom performance patterns
algorithm_groups = {}

# Convert the dataframe to a format where we can use row values as dictionary keys
for idx, row in filtered_df.iterrows():
    # Convert row values to a tuple so it can be used as a dictionary key
    row_values = tuple(row.values)
    
    # Add this algorithm to the appropriate group
    if row_values in algorithm_groups:
        algorithm_groups[row_values].append(idx)
    else:
        algorithm_groups[row_values] = [idx]

# Display the groups
print(f"Found {len(algorithm_groups)} distinct algorithm performance patterns")
for i, (_, algorithms) in enumerate(algorithm_groups.items(), 1):
    print(f"\nGroup {i} ({len(algorithms)} algorithms):")
    for alg in algorithms:
        print(f"  - {alg}")
```

## Complete Axiom Satisfaction

```python
# Find which algorithms passed all test cases for each axiom
# First, get the unique axioms (top level of the column MultiIndex)
axioms = df.columns.get_level_values(1).unique()

# Create a new dataframe to store the results
axiom_satisfaction = pd.DataFrame(index=df.index, columns=axioms)
```

```python
# For each axiom, check if the algorithm passed all subtests
for axiom in axioms:
    # Get all subtests for this axiom by selecting columns where level 1 equals the axiom
    axiom_subtests = df.loc[:, df.columns.get_level_values(1) == axiom]
    # Check if all values are 1 (passed) for each algorithm
    axiom_satisfaction[axiom] = axiom_subtests.all(axis=1)

# Convert boolean values to 1s and 0s for easier reading
axiom_satisfaction = axiom_satisfaction.astype(int)

# Display the results
print("Algorithms that passed ALL test cases for each axiom:")
print(axiom_satisfaction)

# Count how many axioms each algorithm fully satisfies
axiom_counts = axiom_satisfaction.sum(axis=1)
print("\nNumber of axioms fully satisfied by each algorithm:")
print(axiom_counts.sort_values(ascending=False))

# Count how many algorithms fully satisfy each axiom
algorithm_counts = axiom_satisfaction.sum(axis=0)
print("\nNumber of algorithms that fully satisfy each axiom:")
print(algorithm_counts.sort_values(ascending=False))
```
