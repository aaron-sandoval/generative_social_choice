## Load data

```python
from pathlib import Path

data_dir_path = get_base_dir_path() / "data" / "voting_algorithm_evals"
use_latest_data_file = True

if use_latest_data_file:
    # Find the most recently created CSV file
    data_files = list(data_dir_path.glob("*.csv"))
    data_file = max(data_files, key=lambda p: p.stat().st_mtime) if data_files else None
else:
    # Use the hard-coded file
    data_file = data_dir_path / "2025-03-04-204044.csv"
```

```python

# Load the data with the correct MultiIndex structure
df = pd.read_csv(data_file, index_col=0, header=[0,1])
df
```
